```python
from PIL import Image
import os

target_size = (128, 128)
#data_20240213t1633_insulting_only
folder_path = "./data_demove_20240213t1633_insulting_only/"
folder_path2 = "./data_demove_sized_20240213t1633_insulting_only/"
if not os.path.exists(folder_path):
    os.mkdir(folder_path)
if not os.path.exists(folder_path2):
    os.mkdir(folder_path2)
for filename in os.listdir(folder_path):
    filepath = os.path.join(folder_path, filename)
    filepath2 = os.path.join(folder_path2, filename)
    print(filename)
    if filepath.endswith(".jpg"):
        #print('k')
        with Image.open(filepath) as img:
            if img.mode == "P":
                img = img.convert('RGB')
            resized_img = img.resize(target_size)
            resized_img.save(filepath2)
```
